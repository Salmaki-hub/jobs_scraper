WEB SCRAPING USING BS4

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
#from prettytable import PrettyTable
```

```python
#JOB website
page = requests.get("https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=Data+Analyst&txtLocation=Pakistan")
WEB_SCRAPER = BeautifulSoup(page.content,'lxml')
job_pages = WEB_SCRAPER.find_all('li',class_='clearfix job-bx wht-shd-bx')
#print(job_pages)
```

```python
#Trying to get the job titles of all the jobs that are showing on the 1st page
#table = PrettyTable(['Job Title','Comapny','Skills','Date Posted'])

for job_page in job_pages:
    job_date_posted = job_page.find('span',class_='sim-posted').span.text
    if 'few' in job_date_posted:
        job_titles = job_page.find('h2').text.replace(' ','')
        job_company_name = job_page.find('h3',class_='joblist-comp-name').text.replace(' ','')
        job_skills = job_page.find('span',class_='srp-skills').text.replace(' ','')
        
        #table.add_row([job_titles,job_company_name,job_skills,job_date_posted])
        print(f"Job Title :{job_titles.strip()}")
        print(f"Company :{job_company_name.strip()}")
        print(f"Skills :{job_skills.strip()}")
        print(f"Date Posted :{job_date_posted.strip()}")
        print('\n')
        #print(table)
```

```python
#Converting data to a DataFrame and then store in excel file
#Initializing arrays of all three things
title_array=[]
company_array=[]
skills_array=[]
date_array=[]
filter_date=['few','1','2','3']

for job_page in job_pages:
    job_date_posted = job_page.find('span',class_='sim-posted').span.text
    if any(x in job_date_posted for x in filter_date):
        job_titles = job_page.find('h2').text
        job_company_name = job_page.find('h3',class_='joblist-comp-name').text.replace(' ','')
        job_skills = job_page.find('span',class_='srp-skills').text.replace(' ','')
        
        title_array.append(job_titles.strip())
        company_array.append(job_company_name.strip())
        skills_array.append(job_skills.strip())
        date_array.append(job_date_posted.strip())

df=pd.DataFrame({'Job Title':title_array, 'Company Name':company_array, 'Skills':skills_array, 'Date Posted':date_array})
#df
df.to_excel('D:\Salman\Data Analyst\Web_Scraping\Jobs_data.xlsx',encoding='utf-8',index=False)
        
```
